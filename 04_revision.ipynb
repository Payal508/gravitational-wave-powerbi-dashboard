```python
import pandas as pd


df = pd.read_csv(r'../data/gw_events_clean.csv')

print(df.shape)
print(df["commonName"].nunique())
print(df["event_name"].nunique())
```

```python
df["commonName"].value_counts().head(20)
```

```python
import numpy as np

# Ensure numeric
df['snr'] = pd.to_numeric(df['snr'], errors='coerce')
df['false_alarm_rate'] = pd.to_numeric(df['false_alarm_rate'], errors='coerce')

# Scientific quality score
epsilon = 1e-300

df['quality_score'] = (
    np.log(df['snr'] + 1e-9) +
    np.log(1 / (df['false_alarm_rate'] + epsilon))
)

# Keep best-quality version of each event
df_clean = (
    df.sort_values(
        ['commonName', 'quality_score'],
        ascending=[True, False]
    )
    .groupby('commonName', as_index=False)
    .first()
)

print(df_clean.shape)
```

```python
df_clean.to_csv('../data/gw_events_clean.csv', index=False)
```
